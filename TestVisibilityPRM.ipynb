```python
import matplotlib.pyplot as plt
%matplotlib inline

import testsuites
from VisibilityPRM import VisibilityPRMPlanner
```

```python
# Tests
for testEnv in testsuites.testSuite:
    planner = VisibilityPRMPlanner(testEnv.scene, testEnv.limits)
    planner.learn(300)
    solution = planner.findPath(testEnv.start, testEnv.goal, 100000)
    print(solution)
    
    # draw the planning
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1,1,1)
    ax.set_xlim(testEnv.limits[0])
    ax.set_ylim(testEnv.limits[1])
    ax.set_title(testEnv.name)
    planner.visualize(solution, nodeSize=80, ax=ax)
```
