```python
import matplotlib.pyplot as plt
%matplotlib inline

import testsuites
from BasicPRM import BasicPRMPlanner
```

```python
# Tests
for testEnv in testsuites.testSuite:
    planner = BasicPRMPlanner(testEnv.scene, testEnv.limits)
    radius = 5
    numNodes = 300
    planner.learn(radius, numNodes)
    solution = planner.findPath(testEnv.start, testEnv.goal, radius)
    print(solution)
    
    # draw the planning
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1,1,1)
    ax.set_xlim(testEnv.limits[0])
    ax.set_ylim(testEnv.limits[1])
    ax.set_title(testEnv.name)
    planner.visualize(solution, nodeSize=80, ax=ax)
```
